```python
# Importing the dataset
df_fake = pd.read_csv("Fake.csv")
df_true = pd.read_csv("True.csv")

# Preprocessing the data
df_fake['class'] = 0
df_true['class'] = 1

# Combine both datasets
df = pd.concat([df_fake, df_true], ignore_index=True)

# Shuffle the dataframe
df = df.sample(frac=1).reset_index(drop=True)

# Handle missing values and convert float values to strings in the 'text' column
df['text'] = df['text'].fillna('').astype(str)

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['text'])
sequences = tokenizer.texts_to_sequences(df['text'])


# Padding
max_len = 1000
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post', truncating='post')
```

```python

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(padded_sequences, df['class'],test_size=0.25)
```

```python

# LSTM Model
lstm_model = Sequential()
vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 100
lstm_model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len))
lstm_model.add(LSTM(units=64, dropout=0.2, recurrent_dropout=0.2))
lstm_model.add(Dense(1, activation='sigmoid'))
lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Training LSTM Model
lstm_history = lstm_model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))


```

```python
# Extract training and validation loss and accuracy curves
lstm_train_loss = lstm_history.history['loss']
lstm_val_loss = lstm_history.history['val_loss']
lstm_train_accuracy = lstm_history.history['accuracy']
lstm_val_accuracy = lstm_history.history['val_accuracy']
```
